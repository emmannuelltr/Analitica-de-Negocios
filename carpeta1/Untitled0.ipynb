##**1. Descripción del Problema**

Una institución de salud busca implementar un modelo de árbol de decisión para mejorar el diagnóstico preventivo y el pronóstico de enfermedades coronarias en sus pacientes. Para este proceso vamos a utilizar las variables extraídas de la base de datos clínica:

Edad: Indica el número de años que posee el paciente, siendo un factor de riesgo biológico fundamental.

Systolic (Presión Sistólica): Representa la presión máxima que ejerce el corazón cuando late (medida en mmHg).

Diastolic (Presión Diastólica): Indica la presión en las arterias cuando el corazón reposa entre latidos (medida en mmHg).

BMI (Índice de Masa Corporal): Es una medida de la grasa corporal basada en la relación entre el peso y la estatura del individuo.

```python
#Se procede con la carga de las librerías de trabajo}
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
```

```python
#Se procede con la carga de datos de trabajo
nxl="/content/2. HeartDisease.xlsx"
XDB=pd.read_excel(nxl, sheet_name=0)

XD=XDB.iloc[:, [1, 2, 3, 4]]
yd=XDB.iloc[:, [0]]
display(XDB)
```

```python
#Se procede con la implementaciòn del modelo de árbol
from sklearn.metrics import confusion_matrix

mar = DecisionTreeClassifier(criterion="entropy", max_depth=3) # Ajustado a tu reto de 3 niveles
mar.fit(XD, yd)

ydp = mar.predict(XD)   # Pronóstico del modelo sobre los datos históricos

# 2. Se construye la matriz de confusión
cm = confusion_matrix(yd, ydp)
VN = cm[0,0]; FP = cm[0,1]; FN = cm[1,0]; VP = cm[1,1]
print("Matriz de confusión: \n", cm)

# 3. Métricas de desempeño (Basadas en tu ejemplo)
Ex = (VP + VN) / len(XD) # Exactitud: Comportamiento general del modelo
print("Exactitud: ", Ex)

Sen = VP / (VP + FN) # Sensibilidad: Capacidad para detectar enfermos (Disease=1)
print("Sensibilidad (Recall): ", Sen)

Sp = VN / (VN + FP) # Especificidad: Capacidad para detectar personas sanas (Disease=0)
print("Especificidad: ", Sp)

Prec = VP / (VP + FP) # Precisión: De los pronosticados como enfermos, ¿cuántos lo están?
print("Precisión: ", Prec)

PreNeg = VN / (VN + FN) # Precisión Negativa: De los pronosticados como sanos, ¿cuántos lo están?
print("Precisión Negativa: ", PreNeg)
```

```python
from sklearn.tree import export_graphviz #exporta datos a un gráfico
from pydotplus import graph_from_dot_data #crea un gráfico a partir de los datos

vc=["Age","Systolic","Diastolic","BMI"] #Título del árbol
dot_data=export_graphviz(mar, feature_names=vc) #Exportar de números a gráfico
graph=graph_from_dot_data(dot_data) #Crea el gráfico
graph.write_png("Arbol.png")
```

```python
from sklearn.tree import export_graphviz #exporta datos a un gráfico
from pydotplus import graph_from_dot_data #crea un gráfico a partir de los datos

# Títulos de las variables de tu archivo Heart Disease
vc=["Age","Systolic","Diastolic","BMI"]

# Exportar de números a gráfico
dot_data=export_graphviz(mar, feature_names=vc, class_names=['Sano', 'Enfermo'], filled=True, rounded=True)

# Crea el gráfico
graph=graph_from_dot_data(dot_data)

# Guarda el gráfico en formato PDF
graph.write_pdf("Arbol.pdf")
```
